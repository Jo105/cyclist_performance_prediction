## Environment Set-Up

### Load relevant Python Packages

```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# Load your dataframe
df = pd.read_csv('data/pandas_profil.csv')
```

```python
# Then run Pandas Profiling
profile = ProfileReport(df, title="Pandas Profiling Report", explorative=True, minimal=True)
```

```python
# and take a look at the profile, This is quite large so better do it in a separate notebook
profile
```

```python
# Save the profil as .html file
profile.to_file("pandas_profiling/my_pandas_profiling_report.html")
```
